```python
from faker import Faker
import wordcloud
import random
import matplotlib.pyplot as plt
```

# Funções

## Gerar os dados

```python
def gerar_dados(n):
    fake = Faker()
    matriz = []
    for _ in range(n):
        matriz.append([fake.name(),random.randint(0,10)])
    return matriz
```

## Salvar os dados no arquivo

```python
def gravar(mat,arq):
    with open(arq,'w') as f:
        for linha in mat:
            f.write(linha[0] + ',' + str(linha[1]) + '\n')
```

## Obter os dados do arquivo

```python
def ler_arquivo(arq):
    with open(arq,'r') as f:
        linhas = f.readlines()
        linhas = [line.strip().split(',') for line in linhas]
        linhas = [[line[0],int(line[1])] for line in linhas]
    return linhas
```

# Gera novos dados e insere em um arquivo de texto

```python
dados = gerar_dados(100)
gravar(dados,'dados.txt')
```

# Recupera os dados do arquivo para uma matriz

```python
matriz = ler_arquivo('dados.txt')
print(matriz[:5])
```

```python
nomes, valores = zip(*matriz)

#Histograma de propabilidade
plt.hist(valores,density=True)
plt.ylabel("Probabilidade")
plt.xlabel("Valores")
plt.title("Histograma das pontuações")
plt.grid()
plt.show()
```

# Wordcloud de números por extenso

```python
#Dicionário de numeros por extenso
numeros = {0:'zero',1:'um',2:'dois',3:'três',4:'quatro',5:'cinco',6:'seis',7:'sete',8:'oito',9:'nove',10:'dez'}

# Salva os nomes de cada número em uma nova lista
lista_por_extenso = []
for valor in valores:
    lista_por_extenso.append(numeros[valor])

# Gera a wordcloud
text = " ".join(lista_por_extenso)
cloud = wordcloud.WordCloud().generate(text)
plt.figure(figsize=(8,6))
plt.imshow(cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
